```python

#@markdown # Exstensions Download
import os
import subprocess
from IPython.display import clear_output
destination_path = None

extensions = {
    "ext1": {
        "name": "Canvas-Zoom",
        "url": "https://github.com/richrobber2/canvas-zoom",
    },
    "ext2": {
        "name": "3D-open-pose-editor",
        "url": "https://github.com/nonnonstop/3d-open-pose-editor",
    },
    "ext3": {
        "name": "Control-Net",
        "url": "https://github.com/lllyasviel/ControlNet",
    },
    "ext4": {
        "name": "Multydeffusion-Tiling",
        "url": "https://github.com/pkuliyi2015/multidiffusion-upscaler-for-automatic1111"
    }
}

SELECTED_EXTENSION = "Canvas-Zoom" #@param ["All", "Canvas-Zoom", "3D-open-pose-editor", "Control-Net", "Multydeffusion-Tiling"]

if SELECTED_EXTENSION == "All":
  for ext_name, ext_data in extensions.items():
   project_name = ext_data["name"]
   project_url = ext_data["url"]
   print(" [1;32mDownload extension:" + project_name + "...")

   destination_path = '/content/light_colab_expanded/extensions/' + project_name
   subprocess.run(['git', 'clone', project_url, destination_path])
   print(' [1;32mDone!')
   print(' [1;32m---')

  clear_output()

  print(' [1;32mAll extensions have been installed!')

else:
  for ext_name, ext_data in extensions.items():
    if ext_data['name'] == SELECTED_EXTENSION:
     project_name = ext_data["name"]
     project_url = ext_data["url"]
     destination_path = '/content/light_colab_expanded/extensions/' + project_name
     print(" [1;32mDownload extension...")
     subprocess.run(['git', 'clone', project_url, destination_path])
     clear_output()

clear_output()
print(' [1;32mDone!')

if not os.path.exists(destination_path):
  os.makedirs(destination_path)
```
